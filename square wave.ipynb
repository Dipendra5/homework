```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
plt.figure(figsize=[15,4])
L=1
gap=20
x=np.arange(1,20,1/gap)
s=[n for n in range(1,15)]
sm=0
for n in s:
    if n%2!=0:
        y=(4/np.pi)*(1/n)*(np.sin(n*np.pi*x/L))
        sm=sm+y
plt.plot(x,sm)
```
